```python
import matplotlib.pyplot as plt
from sklearn import datasets
import pandas as pd
import numpy as np
#importing modules,packages and libraries
```

```python
Iris=datasets.load_iris()
X = Iris.data[:, :3]
y = Iris.target
#This might give independent and dependent variable lists
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test, y_train, y_test=train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.svm import SVC
model=SVC()
#Gives a support vector classifier,SVC ig
```

```python
model.fit(x_train, y_train)
```

```python
predictions_yhat=model.predict(x_test)
print(predictions_yhat)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

print(confusion_matrix(y_test, predictions_yhat))
```

```python
print(classification_report(y_test, predictions_yhat))
```

```python
#comparative analysis of kernels
from sklearn.metrics import accuracy_score
model_kernel_linear = SVC(kernel='linear')
model_kernel_linear.fit(x_train, y_train)
Yhat_linear = model_kernel_linear.predict(x_test)
print('Accuracy score - Linear Kernel: ', round(accuracy_score(y_test, Yhat_linear),3))

model_kernel_polynomial = SVC(kernel='poly')
model_kernel_polynomial.fit(x_train, y_train)
Yhat_poly = model_kernel_polynomial.predict(x_test)
print('Accuracy score - Poly Kernel: ', round(accuracy_score(y_test, Yhat_poly),3))

model_kernel_rbf= SVC(kernel='rbf')
model_kernel_rbf.fit(x_train, y_train)
Yhat_rbf = model_kernel_rbf.predict(x_test)
print('Accuracy score - RBF Kernel: ', round(accuracy_score(y_test, Yhat_rbf),3))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df1 = pd.read_csv("/content/drive/MyDrive/Copy of diabetes.csv")
df2 = df1.copy()
```

```python
df2.shape
df2.head()
df2.describe()
```

```python
#Checking for missing values.
df2.isnull().values.any()
```

```python
zeroless = ["Glucose","BloodPressure","SkinThickness"]

for column in zeroless:
    df2[column] = df2[column].replace(0, np.NaN)
    mean = int(df2[column].mean(skipna = True))
    df2[column] = df2[column].replace(np.NaN, mean)
```

```python
 #Splitting the dataset into training and testing sets.
from sklearn.model_selection import train_test_split
x = df2.iloc[:, :-2]
y = df2.iloc[:, -1]
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state = 0, test_size = 0.2)
```

```python
from sklearn.svm import SVC

from sklearn.metrics import accuracy_score
model_kernel_linear = SVC(kernel='linear')
model_kernel_linear.fit(x_train, y_train)
Yhat_linear = model_kernel_linear.predict(x_test)
print('Accuracy score - Linear Kernel: ', round(accuracy_score(y_test, Yhat_linear),3))

model_kernel_polynomial = SVC(kernel='poly')
model_kernel_polynomial.fit(x_train, y_train)
Yhat_poly = model_kernel_polynomial.predict(x_test)
print('Accuracy score - Poly Kernel: ', round(accuracy_score(y_test, Yhat_poly),3))

model_kernel_rbf= SVC(kernel='rbf')
model_kernel_rbf.fit(x_train, y_train)
Yhat_rbf = model_kernel_rbf.predict(x_test)
print('Accuracy score - RBF Kernel: ', round(accuracy_score(y_test, Yhat_rbf),3))
```
