```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn import datasets
```

```python
Iris=datasets.load_iris()
X = Iris.data[:, :3]
y = Iris.target
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test, y_train, y_test=train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.metrics import silhouette_score, adjusted_rand_score
from sklearn.metrics import davies_bouldin_score
```

```python
for i in [0.4,0.5,0.6,0.7,0.75,0.8,0.85,0.9]:
    for j in [5,6,7,8,9]:
        from sklearn.cluster import DBSCAN
        clustier = DBSCAN(eps=i, min_samples=j)
        print("this is example with eps=",i,"and min_samples=",j)
        clustier.fit(X)
        labels = clustier.labels_
        n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
        n_noise_ = list(labels).count(-1)
        print(labels)
        print("Estimated number of clusters: %d" % n_clusters_)
        print("Estimated number of noise points: %d" % n_noise_)
        print("Silhouette Coefficient: %0.3f" % silhouette_score(X, labels))
        print("davies_bouldin_score: %0.3f" % davies_bouldin_score(X, labels))
```

```python
#optimal value (approx) of eps is 0.9 and min_samples is 9
```
