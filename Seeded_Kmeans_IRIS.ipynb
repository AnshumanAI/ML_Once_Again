**Import Packages**

```python
#import necessary libraries
from sklearn.cluster import KMeans
import pandas as pd
# import numpy as np
# import matplotlib.pyplot as plt
```

**Import Iris dataset**

```python
#@title Import Iris dataset
# import some data to play with
from sklearn import datasets
iris = datasets.load_iris()
# Take the first two features. We could avoid this by using a two-dim dataset
X = iris.data[:, :2]
y = iris.target
print(X[0:5])
print(y[0:5])
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test, y_train, y_test=train_test_split(X,y,test_size=0.30)
```

```python
#from sklearn.cluster import KMeans  #Importing K-means algorithm
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(x_train)
print("No. of training samples :", len(x_train))
predict_train=kmeans.predict(x_train)
print("Labels of training data :", kmeans.predict(x_train))
```

**Predicting the cluster label of unseen (test) data**

```python
y_kmeans = kmeans.predict(x_test)
print("predicted  label :", y_kmeans)
print("No. of data instances in testing data : ", len(x_test))
```

**Checking the cluster quality in terms of silhouette_score or Silhouette Coefficient. It is a metric used to calculate the goodness of a clustering technique. Its value ranges from -1 to 1. 1: Means clusters are well apart from each other and clearly distinguished**

```python
from sklearn.metrics import silhouette_score
print("K-means Silhouette Score(K=3):", silhouette_score(x_train, predict_train))
```

```python
import matplotlib.pyplot as plt
plt.scatter(x_test[:,0],x_test[:,1], s=50)
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("before applying clustering")
```

```python
plt.scatter(x_test[:,0],x_test[:,1], c=y_kmeans, s=50, cmap='viridis')   #plotting  the results after applying clustering, reference: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html
plt.xlabel('Feature-1')
plt.ylabel('Feature-2')
plt.title("after applying clustering")
```
